```python
import torch
import whisper
from IPython.display import Audio, display

# Check if CUDA is available for GPU usage
use_cuda = torch.cuda.is_available()
device = "cuda" if use_cuda else "cpu"

# Load the Whisper model (large version)
model = whisper.load_model("large", device=device)
```

```python
# Path to the audio file you want to transcribe
audio_path = "/content/test.mp3"  # Replace with your actual file path

# Play the audio
display(Audio(audio_path, autoplay=True)) # Set autoplay=True to play automatically

# Transcribe the audio to text
result = model.transcribe(audio_path)

# Output the transcription
print("Transcription:\n", result["text"])
```
