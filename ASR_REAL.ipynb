```python
# Library Import
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
from IPython.display import Audio, display
import torchaudio

# Choose a Whisper model
model_id = "openai/whisper-large"  # ‚Üê Whisper model

use_cuda = torch.cuda.is_available()
device = "cuda" if use_cuda else "cpu"
torch_dtype = torch.float16 if use_cuda else torch.float32

# Load Whisper model and processor
model = AutoModelForSpeechSeq2Seq.from_pretrained(model_id, torch_dtype=torch_dtype)
model = model.to(device)

processor = AutoProcessor.from_pretrained(model_id)

# Setup pipeline for ASR
pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    device=0 if use_cuda else -1,   # correct device specification
)
```

```python
# Provide the path to the Audio after upload
audio_path = "/content/earhart.mp3"
```

```python
# Play the audio
display(Audio(audio_path, autoplay=True)) # Set autoplay=True to play automatically
```

```python
# Perform ASR and generate transcription
result = pipe(audio_path)
print("ASR Output:")
print(result["text"])
```
